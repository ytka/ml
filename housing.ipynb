# 入力変数と出力変数の切り分け

```python
df.head(3)
```

```python
# df.iloc(行, 列)
df.iloc[0,0]
```

```python
X = df.iloc[:,0:-1]
y = df.iloc[:,-1]
y
```

# モデル構築と検証

```python
from sklearn.linear_model import LinearRegression
```

```python
# モデルの宣言
model = LinearRegression()
```

```python
# モデルの学習
model.fit(X,y)
```

```python
# 検証(決定係数の計算)
model.score(X, y)
```

# 訓練データ(train)と検証データ(test)を分割
例) 受験勉強用10年分の過去問を購入

駄目なケース
- 10年分で勉強（学習）
- 10年分で実力テスト（検証）

良いケース
- 前半の5年分で勉強（学習） <- 訓練データ
- 後半の5年分で実力テスト（検証） <- 検証データ

```python
from sklearn.model_selection import train_test_split
```

```python
# 訓練データと検証データの分割
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.4, random_state=1)
```

```python
# モデルの学習
model.fit(X_train, y_train)
```

```python
# 検証 <- 検証データ
model.score(X_test, y_test)
```

```python
# 検証 <- 訓練データ
model.score(X_train, y_train)
```
